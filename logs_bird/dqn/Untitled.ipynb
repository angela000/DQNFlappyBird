```python
import matplotlib.pyplot as plt
import numpy as np

scores_file = open('./scores.txt', 'r')
scores_str = scores_file.readline().split(" ")
scores_str = scores_str[0:-1]
scores = list(map(eval, scores_str))
scores_file.close()

lost_hist_file = open('./lost_hist.txt', 'r')
lost_hist_list_str = lost_hist_file.readline().split(" ")
lost_hist_list_str = lost_hist_list_str[0:-1]
loss = list(map(eval, lost_hist_list_str))
lost_hist_file.close()

rewards_file = open('./reward.txt', 'r')
rewards_str = rewards_file.readline().split(" ")
rewards_str = rewards_str[0:-1]
rewards = list(map(eval, rewards_str))
rewards_file.close()
```

```python
plt.figure()
plt.plot(loss, '-')
plt.ylabel('loss')
plt.xlabel('time step')
plt.savefig("./1500000_lost_hist_total.png")

plt.figure()
plt.plot(scores, '-')
plt.ylabel('score')
plt.xlabel('episode')
plt.savefig("./1500000_scores_total.png")

plt.figure()
plt.plot(rewards, '-')
plt.ylabel('rewards')
plt.xlabel('episode')
plt.savefig("./1500000_rewards_total.png")
```
